```python
from matplotlib import pyplot as plt
import fastf1
import fastf1.plotting

fastf1.plotting.setup_mpl()

session = fastf1.get_session(2021, 'Spain', 'R')

session.load()
```

```python
ver_laps = session.laps.pick_driver('VER')
ham_laps = session.laps.pick_driver('HAM')
#print(str(ver_laps))

# The rest is just plotting
fig, ax = plt.subplots()
ax.plot(ver_laps['LapNumber'], ver_laps['LapTime'], label='VER')
ax.plot(ham_laps['LapNumber'], ham_laps['LapTime'], label='HAM')
#ax.plot(t2, vCar2, label='SAI')
ax.set_xlabel('Lap Number')
ax.set_ylabel('Time')
#ax.set_title('HAM vs. VER Lap Times, 2021 Spanish GP')
ax.legend()
plt.show()
```
